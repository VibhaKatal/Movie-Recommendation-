```python
separator = ','


df['overview'] = df['overview'].apply(lambda x: separator.join(x) if x is not None else None)
```

```python
newdf = df[['id', 'title', 'release_date', 'tags']]
```

```python
newdf['tags'] = newdf['tags'].apply(lambda x: " ".join(x.split(",")) if isinstance(x, str) else x)
```

```python
newdf
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))


    return ' '.join(y)
```

```python
newdf['tags']=newdf['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=1000,stop_words='english')
```

```python
vectors = cv.fit_transform(newdf['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
outer_batch_size = 1000  
inner_batch_size = 10  
num_vectors = vectors.shape[0]

similarities = []

for i in range(0, num_vectors, outer_batch_size):
    start = i
    end = min(i + outer_batch_size, num_vectors)
    batch_vectors = vectors[start:end]
    
    batch_similarities = []
    for j in range(0, num_vectors, inner_batch_size):
        inner_start = j
        inner_end = min(j + inner_batch_size, num_vectors)
        inner_batch = vectors[inner_start:inner_end]
        inner_batch_similarities = cosine_similarity(batch_vectors, inner_batch)
        batch_similarities.append(inner_batch_similarities)
    
    batch_similarities = np.concatenate(batch_similarities, axis=1)
    
    similarities.append(batch_similarities)

similarity_matrix = np.concatenate(similarities, axis=0)

cv.get_feature_names_out()
```

```python
def recommend(movie):
    movie_index = newdf[newdf['title'] == movie].index[0]
    distances = similarity_matrix[movie_index]
    distances_sorted_indices = distances.argsort()[::-1]
    
    for i in distances_sorted_indices[1:11]:
        print(newdf.iloc[i].title)
```

```python
recommend('Kenali')
```

```python
chunk_size = 1000

num_rows, num_cols = similarity_matrix.shape

with open('similarity_matrix.pkl', 'wb') as f:
    pickle.dump(similarity_matrix.shape, f)
    
    for i in range(0, num_rows, chunk_size):
        start = i
        end = min(i + chunk_size, num_rows)
        chunk = similarity_matrix[start:end, :]
        pickle.dump(chunk, f)
```
